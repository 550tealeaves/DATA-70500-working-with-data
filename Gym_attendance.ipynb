```python
import pandas as pd
import matplotlib.pyplot as plt

# 1) Load your CSV file
# Replace 'your_file.csv' with your actual filename
df = pd.read_csv('daily_gym_attendance.csv')

# Clean up column names by stripping whitespace
df.columns = df.columns.str.strip()

# Peek at the structure (optional but very helpful)
print(df.head())

# 2) Choose your columns for the bar chart
x = df["membership_type"]
y = df["workout_type"]

# 3) Create the bar chart
plt.figure(figsize=(8,5))
plt.bar(x, y)

# 4) Make it readable
plt.xlabel("Membership")
plt.ylabel("Workout")
plt.title("Daily Gym Attendance")
plt.xticks(rotation=45, ha="right")  # tilts labels if they’re long

# 5) Show it
plt.tight_layout()
plt.show()
```

```python
# Clean up column names by stripping whitespace
df.columns = df.columns.str.strip()

# 2) Choose your columns for the bar chart
x = df["age"]
y = df["calories_burned"]

# 3) Create the bar chart
plt.figure(figsize=(7,4))
plt.bar(x, y, color="gold")

# 4) Make it readable
plt.xlabel("Age")
plt.ylabel("Calories burned")
plt.title("Daily Gym Attendance")
plt.xticks(rotation=45, ha="right")  # tilts labels if they’re long

# 5) Show it
plt.tight_layout()
plt.show()
```

```python
# Example data
x = df["calories_burned"]
y = df["workout_duration_minutes"]

plt.figure(figsize=(8,5))
plt.plot(x, y, color="pink")

plt.xlabel("Age")
plt.ylabel("Workout duration mins")
plt.title("Daily Gym Attendance")

plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["calories_burned"], df["workout_duration_minutes"],
            s=10,          # point size
            alpha=0.7)     # transparency

plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["age"], bins=10)

plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution")

plt.show()
```
