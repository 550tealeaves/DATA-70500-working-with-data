# **Basic Data Cleaning**
- Convert data types (year from int to datetime)
- Check for and remove duplicates

```python
# get .info() method to get info about dataset and types
refugee_df.info()
```

```python
# target year col in dataframe and convert it from integer to date/time type
refugee_df['year'] = pd.to_datetime(refugee_df['year'], format="%Y")
```

```python
#check datatypes to make sure conversion worked
refugee_df.dtypes
```

```python
#check for duplicate rows w/ .duplicated() method
# parameter keep=False - shows duplicated values
refugee_df[refugee_df.duplicated(keep=False)]
```

```python
# Use .drop_duplicates() method to remove duplicates
# Use keep='first' to keep the 1st instance of the duplicate of the last
refugee_df = refugee_df.drop_duplicates(keep='first')
```

```python
# Run .duplicated() method to see if duplicates were removed
refugee_df[refugee_df.duplicated(keep=False)]
```

# **Sort columns, groupby columns, and calculations**
- Be mindful of when dataframe converts to series object
- Series object (1D array) cannot do calculations that DataFrame (2D tabular structure) can
- Use .to_frame() method & reset_index() method to convert series back to dataframe

```python
# sort columns using .sort_values() method & by= (parameter) and col name
# example sorts dataframe by arrivals in descending order
refugee_df.sort_values(by='arrivals', ascending=False)
```

```python
# use .groupby() method to group data
# see breakdown of # arrivals by country of origin
refugee_df.groupby('origin')
```

```python
# Groupby origin
# Summing # of arrivals from each country of origin
# Sort values in descending order
# Take the top 20 values
refugee_df.groupby('origin')['arrivals'].sum().sort_values(ascending=False)[:20]
```

```python
# Covert series object into dataframe (series doesn't let us do calculations)
# Use to_frame & reset_index
ref_sum_df = refugee_df.groupby('origin')['arrivals'].sum().sort_values(ascending=False)[:20]
ref_sum_df = ref_sum_df.to_frame().reset_index() #converts from series to data frame & resets index
ref_sum_df
```

# **Data Visualizations**
## Use .plot() method and kind= (parameter) to determine what plots to apply
- Viz helps easily translate data insights to users

```python
# Create bar chart using .plot() method
# \n is  line break
ref_sum_df.plot(kind='bar', x='origin', y='arrivals', color='green', title='Total number of refugee arrivals in the U.S. \n by country of origin')
```
